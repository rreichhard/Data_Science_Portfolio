```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import codecs
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
with codecs.open('avatar_fulltextjson.json', 'r', 'utf-8-sig') as data:
    text = json.load(data)
```

```python
mask = np.array(Image.open('twi.jpg'))
```

```python
# wc = WordCloud(stopwords = STOPWORDS, mask=mask, background_color='white',
#               max_words = 100, max_font_size = 500, random_state = 42,
#               width = 15000,
#               height = 10000)
```

```python
wc = WordCloud(stopwords = STOPWORDS, mask=mask, background_color='white', colormap = 'Dark2',
              max_words = 250, max_font_size = 1000, random_state = 42,
              width = mask.shape[1],
              height = mask.shape[0])
```

```python
wc.generate(text)
plt.figure(figsize=[10,10])
plt.imshow(wc, interpolation = "bilinear")
plt.axis('off')
plt.savefig('atla_wc.jpg')
plt.show()
```
