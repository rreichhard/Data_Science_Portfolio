```python
rnrhof_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_Rock_and_Roll_Hall_of_Fame_inductees', match='Year')
```

```python
len(rnrhof_table)
```

```python
df = rnrhof_table[0]
```

```python
df.drop(columns = ['Image'], inplace=True)
```

```python
df.rename(columns={'Inducted by[5]': 'Inducted By', 'Inducted members':'Inducted Members', 'Year':'Induction Year'}, inplace=True)
```

```python
df.dtypes
```

```python
df['Inducted By'] = df['Inducted By'].astype(str)
df['Inducted Members'] = df['Inducted Members'].astype(str)
df['Name'] = df['Name'].astype(str)
```

```python
# df.loc[:,df['Inducted Members']].replace(np.nan,"Solo Artist", regex=True, inplace=True)
```

```python
df.tail(50)
```

```python
df['Inducted By'] = df['Inducted By'].apply(lambda x: x.replace("["," "))
```

```python
df['Inducted By'] = df['Inducted By'].apply(lambda x: x.replace("]"," "))
```

```python
df['Inducted By'] = df['Inducted By'].apply(lambda x: x.replace("N2",""))
df['Inducted By'] = df['Inducted By'].apply(lambda x: x.replace("N3",""))
```

```python
df['Inducted Members'] = df['Inducted Members'].apply(lambda x: x.replace("["," "))
df['Inducted Members'] = df['Inducted Members'].apply(lambda x: x.replace("]"," "))
df['Inducted Members'] = df['Inducted Members'].apply(lambda x: x.replace("22"," "))
```

```python
df['Name'] = df['Name'].apply(lambda x: x.replace("["," "))
df['Name'] = df['Name'].apply(lambda x: x.replace("]"," "))
df['Name'] = df['Name'].apply(lambda x: x.replace("N2"," "))
df['Name'] = df['Name'].apply(lambda x: x.replace("137"," "))
```
