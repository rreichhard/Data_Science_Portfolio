```python
# Import necessary libraries

import pandas as pd
from translate import Translator
import string
```

```python
# Read the CSV and look at the first 10 rows.


df = pd.read_csv('tracks.csv')
df.head(10)
```

```python
# Checking to see the datatype of each column

print(df.dtypes)
```

# <font color=red> Data Cleaning Step Number 1 </font>

```python
# The release date of the track is showing as an object and I would like it to be in datetime format

df['release_date'] = df['release_date'].astype('datetime64[ns]')
df['name'] = df['name'].astype(str)
df['artists'] = df['artists'].astype(str)
```

```python
print(df.dtypes)
```

```python
# Checking to make sure release_date is still showing in a readable format

df.head(10)
```

# <font color=red> Data Cleaning Step Number 2 </font>

```python
# the artist name and artist id columns are all showing as encased in a [''] format. 
# I'd like to remove that so the names and id's are similar to the id and name of the tracks
# str.lstrip removes leading characters and str.rstrip removes trailing characters

df['artists'] = df['artists'].str.lstrip("['")
df['artists'] = df['artists'].str.rstrip("']")
df['id_artists'] = df['id_artists'].str.lstrip("['")
df['id_artists'] = df['id_artists'].str.rstrip("']")
df['artists'] = df['artists'].str.replace("', '"," ")
```

```python
df.tail(100)
```

# <font color=red> Data Cleaning Step Number 3 </font>

```python
# Renaming columns to make more sense

df2 = df.rename(columns={"id":"track_id","name":"track_name","artists":"artist_name","id_artists":"artist_id"})
```

# <font color=red> Data Cleaning Step Number 4 </font>

```python
# My original intention for this cleaning step was to translate track titles and artist names into English
# After a great deal of troubleshooting I am still unable to get an actual translation
# I would love some pointers on how to get this to work instead of the code just replicating the original name

translator = Translator(to_lang= 'en')
df2['English'] = df2['track_name'].apply(lambda x: translator.translate(x))
df2.tail()
```

```python
# Working Step#3 replacing punctuation in the artist and track titles and then making them all lower case

df2['track_name'] = df2['track_name'].str.replace(r'[^\w\s]+',"")
df2['artist_name'] = df2['artist_name'].str.replace(r'[^\w\s]+',"")
df2['track_name'] = df2['track_name'].apply(lambda x: x.lower())
df2['artist_name'] = df2['artist_name'].apply(lambda x: x.lower())
```

```python
df2.tail(100)
```

# <font color=red> Data Cleaning Step Number 5 </font>

```python
# checking to see if there are duplicates/outliers
# a sum of zero means that there are no duplicate track_id's in the dataset

df2.duplicated(subset = 'track_id').sum()
```

```python
# obtaining basic descriptive statistics for the numerical attributes to ensure there are no outliers

df2.describe()
```
